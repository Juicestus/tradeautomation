```python
import json
    
def load():
    with open("sbosc_results.json", 'r') as f:
        s = f.read()
#         ss = s.split('}{')
#         return json.loads('{' + ss[-1])
        return json.loads(s)
    
data = load()
```

```python
def zipdata(data):
    zl = []
    for _len, submap in data.items():
        for sig_len, tpr in submap.items():
            zl.append((_len, sig_len, tpr))
    return zl

#print(zipped)

def unzip(zipped):
    lens, sig_lens, tprs = np.zeros((3, len(zipped)))
    for i, t in enumerate(zipped):
        lens[i] = t[0]
        sig_lens[i] = t[1]
        tprs[i] = t[2]
    return lens, sig_lens, tprs 

```

```python
zipped = zipdata(data)
lens, sig_lens, tprs = unzip(zipped)

fig = plt.figure(figsize=(30, 30))
ax1 = fig.add_subplot(121, projection='3d')

bot = np.zeros_like(tprs)
for i, _ in enumerate(tprs):
    if _ < 0:
        tprs[i] = 0
        

width = .5
depth = .5

ax1.bar3d(lens, sig_lens, bot, width, depth, tprs, shade=True, zsort='average')
ax1.set_title('returns(_len, sig_len)')
```

```python
import math

a = []
def pa(s, v):
    print(s + ' = ', v)
    a.append((s,v))
    
pa('π^2', math.pi ** 2)
pa('e^2', math.e ** 2)
pa('eπ ', math.e * math.pi)
pa('π^e', math.pi ** math.e)
pa('e^π', math.e ** math.pi)
pa('π^π', math.pi ** math.pi)
pa('e^e', math.e ** math.e)

print('')
o = ' '.rjust(4)
for kc, vc in a:
    o += kc.rjust(10)
o += '\n'
for kr, vr in a:
    o += kr + " "
    for kc, vc in a:
        if kr == kc:
            o += ' '.rjust(10)
        else:
            o += ('% 6.2f' % (vc * vr)).rjust(10)
    o += '\n'
print(o)
    
    
```
